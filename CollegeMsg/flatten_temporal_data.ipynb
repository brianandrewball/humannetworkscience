# Flatten out the temporal dimension

All this notebook does is take in the college message data and removes the temporal dimension

(u, v, t) ---> (u, v)

It saves a new, flattened dataset

(I suspect this isn't even necessary)

```python
import collections as col
input_file_path  = "CollegeMsg.txt"
output_file_path = "CollegeMsg-Flattened.txt"


```

```python
with open(input_file_path, "r") as indata, open(output_file_path, "w") as outdata:
    for line in indata:
        if line.strip() and line[0] != "#":  # Skip empty lines and comments
            u, v, _ = line.strip().split()  # Split each line into parts and ignore the timestamp
            outdata.write(f"{u} {v}\n")  # Write only the user IDs u and v
```
